# Import Libraries

```python
# For saving and starting the .mp3 file
import os
# For extracting the text from wikipedia
import wikipedia
# For Text-to-Speech conversion
from gtts import gTTS
# For displaying the Audio
from IPython.display import Audio, display
```

# Generating a Long summary in English

```python
wikipedia.set_lang('en')
```

```python
results = wikipedia.summary('Lord Shiva', sentences = 10)
```

```python
results
```

# Generating a Short Summary in English

```python
short_sum = wikipedia.summary('Parvati', sentences = 5)  # Changing the number of sentences
```

```python
short_sum
```

# Hindi Summarization

```python
wikipedia.set_lang('hi')
```

```python
results_hindi = wikipedia.summary('Durga', sentences = 5)
```

```python
results_hindi
```

# Text To Speech 
- GTTS stands for Google Text-to-Speech. It is a service provided by Google that allows users to convert text into spoken words.

```python
myobj = gTTS(text = results, lang = "en", slow = False)
```

```python
obj_two = gTTS(text = short_sum, lang = "en", slow = False)
```

```python
obj_three = gTTS(text = results_hindi, lang = "hi", slow = False)
```

### Creating mp3 files

```python
myobj.save("wikisum.mp3")
```

```python
obj_two.save("shortsum.mp3")
```

```python
obj_three.save("hindisum.mp3")
```
